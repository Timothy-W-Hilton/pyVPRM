```python
%matplotlib inline
from pyVPRM.lib.fancy_plot import *
import xarray as xr
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import cartopy.crs as ccrs
from pyVPRM.lib.sat_manager import satellite_data_manager
```

# Vegetation Map on MODIS Grid

```python
veg_map_modis = xr.open_dataset('/home/b/b309233/software/VPRM_preprocessor/examples/wrf_preprocessor/out/veg_map_on_modis_grid_1_1.nc')
```

```python
cmap = plt.cm.inferno_r
cmaplist = [cmap(i) for i in range(cmap.N)]
# cmaplist[0] = '#c4c4c4'
cmap = mpl.colors.LinearSegmentedColormap.from_list(
    'Custom cmap', cmaplist, cmap.N)
bounds = np.linspace(0, 1, 11)
norm = mpl.colors.BoundaryNorm(bounds, cmap.N)

fig = plt.figure(111, figsize = figsize(1.2, ratio=5))
for c,i in enumerate(np.arange(1,9)[::-1]):
    ax = fig.add_axes((0, c*0.1+0.04,1.0, 0.07), projection=ccrs.PlateCarree())
    cb= veg_map_modis.sel({'vprm_classes': i})['veg_1'].plot.pcolormesh(cmap=cmap, ax=ax,
                                                                             x='lon', y='lat', vmin=0, vmax=1.0,
                                                                             add_colorbar=False)
    ax.set_title('Vegetation Class {}'.format(i))
    axins = inset_axes(ax,
                       width="5%",  
                       height="90%",
                       loc='center right',
                       borderpad=-5
                       )
    gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                      linewidth=1.5, color='gray', alpha=0.5, linestyle='--')
    cbar = fig.colorbar(cb, cax=axins)
    cbar.ax.get_yaxis().labelpad = 15
    cbar.ax.set_ylabel('Vegetation Fraction', rotation=270)

    
```
